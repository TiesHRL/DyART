```python
from skopt.plots import plot_objective
from skopt.space import Space
from skopt.utils import create_result
import pandas as pd
import matplotlib.pyplot as plt
from skopt import Optimizer
from skopt.space import Integer, Real


# Define the search space
space = Space([
    Real(min(df['max_depth']), max(df['max_depth']), name='max_depth'),
    Real(min(df['max_weight']), max(df['max_weight']), name='max_weight'),
    Real(min(df['min_size']), max(df['min_size']), name='min_size'),  # example range, adjust as necessary
    Integer(min(df['p']), max(df['p']), name='p')
])

# Create an optimizer object
opt = Optimizer(space)

# Assume that df['target'] are the function values and df.drop('target', axis=1) are the points
for i in range(len(df1)):
    opt.tell(df1.drop('target', axis=1).iloc[i].tolist(), -df1['target'].iloc[i])

# Use plot_objective
plot_objective(opt.get_result())
plt.show()
```
